# Train- test split

```python
#this is used for learning purpose
import numpy as np
 #seed is used so that by shuffling each time, all data points do not come to the train set.the train set value will be fixed
def train_test_split(data, test_ratio):
    np.random.seed(42)
    shuffled = np.random.permutation(len(data))
    test_set_size = int((len(data) * test_ratio))
    test_indices = shuffled[:test_set_size]
    train_indices = shuffled[test_set_size:]
#     print(test_set_size)
#     print(test_indices)
#     print(train_indices)
    return data.iloc[train_indices], data.iloc[test_indices]


#data.iloc is used to select row and column by numbers
#by doing shuffle, with each run, new data will be used, shuffled data will be used.
#train_test_split function is already implemented in the scikit-learn
```

```python
# train_set, test_set = train_test_split(housing, 0.2)
# print(f"Rows in train set: {len(train_set)} \n Rows in test set: {len(test_set)}\n")
#we want that our train set and test set represent all cases of population, for this reason, stratified shuffle is used
```

```python
from sklearn.model_selection import train_test_split
train_set, test_set = train_test_split(housing, test_size = 0.2, random_state = 42)
print(f"Rows in train set: {len(train_set)} \n Rows in test set: {len(test_set)}\n")
```

```python
from sklearn.model_selection import StratifiedShuffleSplit
split = StratifiedShuffleSplit(n_splits =1, test_size = 0.2, random_state = 42) #CHAS er respect e stratified shuffle shift hobe
for train_index, test_index in split.split(housing, housing['CHAS']):
    strat_train_set = housing.loc[train_index]
    strat_test_set = housing.loc[test_index]
```

```python
# There is 3 option to deal with the missing attributes
# 1. get rid of the missing columns
# 2. get rid of the whole attribute
# 3. set the value to some value (0, mean or median)
```

```python
a = housing.dropna(subset=["RM"]) #option 1
#dropna drops the null values from the dataset
a.shape
```

```python
housing.drop("RM", axis=1).shape #option 2
#there is no RM colmn and original housing dataframe will remain the same
```
